# Introduction

Recently, our class has been engaging in discussions about AI's potential to do nearly any job. I've tried not to think too much about this or my brain will go numb, but I was on a call with a Penn Undergraduate Pre-Grad School advisor and wondered: could AI do this woman's job as a college counselor? She not only had personal relationships with many different universities, but her wealth of knowledge spanned far beyond basic "google-able" information I would be able to find on my own. Thus, the idea for this blog post was born: let's see how well ChatGPT functions as a college advisor for hypothetical high school students.

I decided to start off my exploration simply asking ChatGPT what schools students should apply to given their preferences. Knowing the general information and colloquial "vibes" of many colleges was going to be critical for the platform's ability to work as a college counselor, so this seemed like an important place to begin.

## What should be the topic of my personal essay?

Another critical part of a college advisor's role is to help students write their personal statements for the Common Application (colloquially called the Common App). Real counselors don't normally tell students exactly what experience or theme to write about, so I didn't expect this of ChatGPT, I just asked it to give me advice regarding how to go about thinking of ideas.

### Prompt:
What should I write my Common App personal essay about?

### Response: 

Choosing a topic for your Common App personal essay can be a deeply personal decision, but here are a few potential approaches to consider:

Significant Life Experience: Reflect on a moment or experience that had a profound impact on your life. This could be overcoming a personal challenge, a cultural experience that broadened your perspective, or a moment of personal growth.

Passion or Interest: Write about something you're truly passionate about. Whether it's a hobby, a field of study, or a cause you deeply care about, sharing your enthusiasm can give admissions officers insight into who you are beyond your academic achievements.

Identity and Background: Explore how your identity, background, or cultural heritage has shaped your perspective and influenced your goals. This could include discussing your family background, cultural traditions, or experiences as a member of a marginalized community.

Failure and Resilience: Reflect on a time when you faced failure or adversity and how you responded to it. Admissions officers are often interested in seeing how applicants handle setbacks and challenges, as it demonstrates resilience and maturity.

Personal Growth and Reflection: Discuss how you've evolved as a person over time. This could involve reflecting on lessons learned from past experiences, changes in your worldview, or moments of self-discovery.

Impactful Relationships: Write about a person who has had a significant influence on your life. This could be a family member, teacher, mentor, or friend. Explore how this relationship has shaped your values, aspirations, or personal development.

Unique Perspective or Experience: Share a story or perspective that sets you apart from other applicants. This could be an unusual hobby, an unconventional life experience, or a unique talent or skill.

Ultimately, the most important thing is to choose a topic that feels authentic to you and allows you to showcase your personality, values, and strengths. Once you've selected a topic, focus on crafting a compelling narrative that engages the reader and provides insight into who you are as an individual. Don't be afraid to be vulnerable and introspective, as this can often lead to the most impactful essays.

### Prompt: 
How should I personally go about choosing my topic?

### Response:

Here are some questions to help spark ideas for your Common App personal essay:

1. **What are your most defining moments?** Think about experiences that have shaped who you are today. These could be positive moments of achievement, challenging times of adversity, or moments of personal growth.

2. **What are you passionate about?** Consider your interests, hobbies, and values. What topics or activities energize you and make you feel alive? Reflect on why these passions are meaningful to you and how they contribute to your identity.

3. **What challenges have you overcome?** Reflect on obstacles you've faced and how you've persevered. This could include personal struggles, academic challenges, or setbacks in your extracurricular pursuits.

4. **How has your background influenced you?** Explore how your family, cultural heritage, or upbringing has shaped your perspective and values. Consider the traditions, values, and experiences that have had a significant impact on who you are today.

5. **What are your goals and aspirations?** Think about your dreams for the future and what drives you to pursue them. Consider both your short-term and long-term goals, as well as the motivations behind them.

6. **Who has had a significant impact on your life?** Reflect on the people who have influenced you the most, whether they're family members, teachers, mentors, or friends. Consider how these relationships have shaped your values, beliefs, and aspirations.

7. **What makes you unique?** Consider your individual strengths, talents, and quirks. What sets you apart from others, and how do these qualities contribute to your identity?

Take some time to brainstorm and jot down your thoughts in response to these questions. You might find that certain ideas resonate with you more strongly than others. Once you've identified potential topics, consider which one allows you to convey the most about who you are as a person and which one you feel most passionate about sharing with admissions officers.


### Prompt: 
Should I write about breaking my leg before the big game?


### Response:

Writing about breaking your leg before a big game could potentially make for a compelling essay, but it's important to approach the topic thoughtfully and consider how it reflects your character, values, and aspirations. Here are some points to consider before deciding to write about this topic:

1. **Reflect on the Impact:** Think about how breaking your leg affected you physically, emotionally, and mentally. Did it challenge you to overcome adversity, test your resilience, or change your perspective on life? Reflecting on the impact of the experience can help you convey its significance to the reader.

2. **Focus on Personal Growth:** Consider how you responded to the setback and what you learned from the experience. Did it teach you valuable lessons about perseverance, teamwork, or resilience? Admissions officers are often interested in seeing how applicants grow and develop in the face of challenges.

3. **Connect to Your Goals and Aspirations:** Tie the experience back to your broader goals and aspirations. How did overcoming this challenge shape your future ambitions or influence your values and priorities? Connecting the narrative to your larger journey can help make the essay more meaningful and cohesive.

4. **Avoid Clichés:** Be mindful of clichés or overly dramatic storytelling. While breaking your leg before a big game is certainly a significant event, try to approach the topic with authenticity and nuance. Focus on sharing your genuine thoughts and feelings rather than relying on clichéd narratives.

5. **Consider Alternative Topics:** If you feel that breaking your leg before the big game is too predictable or doesn't fully capture who you are as a person, consider exploring alternative topics that showcase different facets of your personality, interests, or experiences.

Ultimately, whether you choose to write about breaking your leg before the big game depends on how effectively you can convey its significance and relevance to your personal narrative. If you can articulate the lessons learned, personal growth, and connections to your goals in a compelling and authentic way, it could make for a strong essay. However, if you feel that there are other experiences that better showcase your character and values, it may be worth exploring those topics instead.


::: {.callout-tip}

## Conclusion 

This part of my experiment was really what I was most curious about, but frankly, it felt like ChatGPT took the easy way out in most cases. It simply regurgitated the same descriptions of and reasoning for lots of schools, and the responses didn't feel particularly tailored to my questions. I think this is because the LLM doesn't want to lead users to think that its word is law by promising acceptance into any school. While this is accurate to real-life college counselors, humans have general experience and knowledge that can lead them to say "yeah, your odds are higher because of x, y, and z," or "let's consider pivoting what schools we look at." It was really interesting that the platform noted how FGLI or minority background could be helpful in college admissions given the topic's somewhat controversial nature especially given the Supreme Court's decision about affirmative action. 

Overall, I think that LLMs would be most helpful in generating ideas and offering high-level support for students about to embark on their college application journey; I still believe that human experience and niche understanding of the field beat out anything AI can do at this point.
